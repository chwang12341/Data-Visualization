```python
import plotly.graph_objs as go
import plotly.offline as pof
import numpy as np

## 
pof.init_notebook_mode(connected = True)

## data 1
## 2000
x = np.random.randn(2000)
y = np.random.randn(2000)

data1 = go.Scatter(
    x = x,
    y = y,
    marker = {'color': 'blue', 'symbol': 104, 'size': 4},
    mode = 'markers',
    name = 'data1'
)

## data2
## 1000

x1 = np.random.randn(1000)
y1 = np.random.randn(1000)

data2 = go.Scatter(
    x = x1,
    y = y1,
    marker = {'color': 'red', 'size': 4},
    mode = 'markers',
    name = 'data2'
)

## data1data2
data = [data1, data2]

## 
layout = go.Layout(
    title = "Plotly Scatter",
    xaxis = {'title': 'x'},
    yaxis = {'title': 'y'},
    font = {
        'size': 28,
        'family': 'serif',
        'color': '#D9006C'
        
    })  

## 
figure = go.Figure(data = data, layout = layout)

## 
pof.iplot(figure, filename = 'Plotly-Scatter', show_link = True, link_text = "Plotly Links", image_height = 800, image_width = 900)

```

```python
import plotly.graph_objs as go
import plotly.offline as pof

pof.init_notebook_mode(connected = True)

## 
## go.Scatter()
## marker: , symbol - 
## mode: text, lines, markers: text+lines  markers+lines  markers + lines + text
## text: 
## name: 

data1 = go.Scatter(
    x = [2, 7, 9],
    y = [18, 16, 8],
    marker = {'color': 'blue', 'symbol': 104, 'size': 20},
    mode = "markers+lines+text",
    text = ["First", "Second", "Third"],
    name = 'data1'
    
)


data2 = go.Scatter(
    x = [1, 7, 8],
    y = [2, 18, 20],
    marker = {'color': 'red', 'symbol': 106, 'size': 20},
    mode = "markers+lines",
    text = ["First", "Second", "Third"],
    name = 'data2'
    
)


## 
layout = go.Layout(title = "Plotly Scatter", xaxis = {'title': 'x'}, yaxis = {'title': 'y'})

## data
figure = go.Figure(data = [data1, data2], layout = layout)

## 
pof.iplot(figure, filename = 'First Plotly', show_link = True, link_text = "Plotly Links", image = 'png', image_height = 800, image_width = 900)

```

```python
import plotly.graph_objs as go
import plotly.offline as pof
import numpy as np

## 
pof.init_notebook_mode(connected = True)

## data 1
## 100
x = np.random.randn(100)
y = np.random.randn(100)

data1 = go.Scatter(
    x = x,
    y = y,
    marker = {'color': 'blue', 'symbol': 104, 'size': 4},
    mode = 'markers',
    name = 'markers'
)

## data2
x1 = [v for v in range(-2, 4)]
y1 = np.random.randn(10) - 5

data2 = go.Scatter(
    x = x1,
    y = y1,
    marker = {'color': 'red', 'size': 4},
    mode = 'markers',
    name = 'lines'
)


## data3
x2 = [v for v in range(-2, 4)]
y2 = np.random.randn(10) - 10

data3 = go.Scatter(
    x = x2,
    y = y2,
    marker = {'color': 'yellow', 'size': 10},
    mode = 'lines+markers',
    name = 'lines+markers'
)

## data4
x3 = [v for v in range(-2, 4)]
y3 = np.random.randn(10) - 20

data4 = go.Scatter(
    x = x3,
    y = y3,
    marker = {'color': 'black', 'size': 10},
    mode = 'lines+markers+text',
    text = ["First", "Second", "Third", "Forth", "Fifth", "Sixth"],
    name = 'lines+markers+text'
)



## data1data2
data = [data1, data2, data3, data4]

## 
layout = go.Layout(
    title = "Plotly Scatter",
    xaxis = {'title': 'x'},
    yaxis = {'title': 'y'},
    font = {
        'size': 28,
        'family': 'fantasy',
        'color': '#D9006C'
        
    })  

## 
figure = go.Figure(data = data, layout = layout)

## 
pof.iplot(figure, filename = 'Plotly-Scatter', show_link = True, link_text = "Plotly Links", image_height = 800, image_width = 900)

```

```python
import plotly.graph_objs as go
import plotly.offline as pof
import numpy as np

## 
pof.init_notebook_mode(connected = True)

## data 1
## 2000
x = np.random.randn(2000)
y = np.random.randn(2000)

data1 = go.Scatter(
    x = x,
    y = y,
    marker = {'color': 'blue', 'symbol': 104, 'size': 4},
    mode = 'markers',
    name = 'data1'
)

## data2
## 1000

x1 = np.random.randn(1000)
y1 = np.random.randn(1000)

data2 = go.Scatter(
    x = x1,
    y = y1,
    marker = {'color': 'red', 'size': 4},
    mode = 'markers',
    name = 'data2'
)

## data1data2
data = [data1, data2]

## 
layout = go.Layout(
    title = "Plotly Scatter",
    xaxis = {'title': 'x'},
    yaxis = {'title': 'y'},
    font = {
        'size': 28,
        'family': 'serif',
        'color': '#D9006C'
        
    },
    showlegend = True,
    legend = {
        'x': 0.6,
        'yanchor': 'middle',
        'bgcolor': 'pink',
        'bordercolor': 'black'
        
    } ,
    hoverlabel = {
        'bgcolor': 'yellow',
        'bordercolor': 'black',
        'namelength': -1,
        'font': {'color': 'brown'}
    }
    )  

## 
figure = go.Figure(data = data, layout = layout)

## 
pof.iplot(figure, filename = 'Plotly-Scatter', show_link = True, link_text = "Plotly Links", image_height = 800, image_width = 900)

```

```python
## 
import plotly.graph_objs as go
from plotly.subplots import make_subplots

## 2X2
fig = make_subplots(rows = 2, cols = 2)

## 
## 1
fig.add_trace(
    go.Scatter(x = [8, 2, 6, 10, 14, 18], y = [18, 14, 12, 20, 24, 28], mode = 'markers', name = 'Scatter'),
    row = 1,
    col = 1
)


## 2
fig.add_trace(
    go.Scatter(x = [8, 2, 6, 10, 14, 18], y = [18, 14, 12, 20, 24, 28], mode = 'lines', name = 'Lines'),
    row = 1,
    col = 2
)

## 3
fig.add_trace(
    go.Bar(x = [8, 2, 6, 10, 14, 18], y = [18, 14, 12, 20, 24, 28], name = 'Bar'),
    row = 2,
    col = 1
)

## 1
fig.add_trace(
    go.Histogram(x = [8, 2, 6, 10, 14, 18], y = [18, 14, 12, 20, 24, 28], name = 'Histogram'),
    row = 2,
    col = 2
)

## 
fig.show()
```
